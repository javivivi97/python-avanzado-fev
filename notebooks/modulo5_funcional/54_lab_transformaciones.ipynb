## 1Ô∏è‚É£ Escenario

Tenemos una lista de **ventas diarias** en formato `(vendedor, importe)`.

```python
ventas = [
    ('Ana', 1200), ('Luis', 800), ('Carlos', 1500), ('Ana', 700), ('Luis', 950), ('Ana', 300)
]
```

Queremos obtener informaci√≥n como:
- Total de ventas globales.
- Total por vendedor.
- Filtrar vendedores con media superior a 1000‚Ç¨.
- Verificar si **todos** los vendedores superaron un umbral m√≠nimo semanal.

````python
ventas = [
    ('Ana', 1200), ('Luis', 800), ('Carlos', 1500), ('Ana', 700), ('Luis', 950), ('Ana', 300)
]
````

---
## 2Ô∏è‚É£ üß© Ejercicio 1 ‚Äî Total global de ventas

Usa `reduce()` para calcular el **total global** de ventas.

üí° *Pista:* suma los segundos elementos de cada tupla (`importe`).

````python
# Escribe aqu√≠ tu c√≥digo para calcular el total global...

from functools import reduce

total_global = reduce(lambda x, y: x + y[1], ventas, 0)

print(f'Total de ventas: {total_global} ‚Ç¨')
````

### ‚úÖ Soluci√≥n propuesta

````python
from functools import reduce
total = reduce(lambda acc, v: acc + v[1], ventas, 0)
print('üí∞ Total global de ventas:', total, '‚Ç¨')
````

---
## 3Ô∏è‚É£ üß© Ejercicio 2 ‚Äî Total de ventas por vendedor

Genera un diccionario con el total acumulado de cada vendedor.

üí° *Pista:* puedes usar comprensi√≥n de diccionario junto con `filter()` o un bucle `set()` sobre los nombres.

````python
# Escribe aqu√≠ tu implementaci√≥n...
from collections import defaultdict

ventas_vendedor = defaultdict(int)

for vendedor, cantidad in ventas:
    ventas_vendedor[vendedor] += cantidad

ventas_vendedor
````

### ‚úÖ Soluci√≥n propuesta

````python
vendedores = {v[0] for v in ventas}
totales = {v: sum(x[1] for x in ventas if x[0] == v) for v in vendedores}
print('Totales por vendedor:', totales)
````

---
## 4Ô∏è‚É£ üß© Ejercicio 3 ‚Äî Filtrar vendedores destacados

Queremos los vendedores cuya **media de ventas** sea superior a 1000‚Ç¨.

üí° *Pista:* usa `map()` o comprensi√≥n de diccionario con una condici√≥n.

````python
# Escribe aqu√≠ tu filtro de destacados...

vendedores_mas_mil = {vend: ventas for vend, ventas in ventas_vendedor.items() if ventas >1500}
vendedores_mas_mil
````

````python
# Vendedores con venta media > 1000

import statistics

ventas_acumuladas = defaultdict(list)

for vendedor, venta in ventas:
    ventas_acumuladas[vendedor].append(venta)

media_ventas_mas_1000 = {vend: statistics.mean(ventas_acumuladas[vend]) for vend in ventas_acumuladas if statistics.mean(ventas_acumuladas[vend]) > 1000}
media_ventas_mas_1000
````

### ‚úÖ Soluci√≥n propuesta

````python
promedios = {
    v: (sum(x[1] for x in ventas if x[0] == v) / len([x for x in ventas if x[0] == v]))
    for v in vendedores
}
destacados = {v: p for v, p in promedios.items() if p > 1000}
print('üèÖ Vendedores destacados:', destacados)
````

## 5Ô∏è‚É£ üß© Ejercicio 4 ‚Äî Verificar rendimiento global

Usa `all()` para comprobar si **todos** los vendedores superan un umbral m√≠nimo (por ejemplo, 500‚Ç¨ en media semanal).

üí° *Pista:* usa el diccionario de `promedios` anterior.

````python
promedios
````

````python
# Escribe aqu√≠ tu verificaci√≥n con all()...

minimo_alcanzado = all(promedios[vendedores] > 500 for vendedores, _ in promedios.items())
minimo_alcanzado
````

### ‚úÖ Soluci√≥n propuesta

````python
rendimiento_ok = all(media > 500 for media in promedios.values())
print('¬øTodos los vendedores cumplen?', rendimiento_ok)
````

### ‚úÖ Soluci√≥n propuesta

````python
nombres, medias = list(promedios.keys()), list(promedios.values())
resumen = sorted(zip(nombres, medias), key=lambda x: x[1], reverse=True)
print('üìä Ranking de vendedores:')
for n, m in resumen:
    print(f'{n:10} ‚Üí {m:.2f}‚Ç¨')
````
