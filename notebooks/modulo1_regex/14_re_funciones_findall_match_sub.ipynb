# üß© 1.4 ‚Äî Funciones del m√≥dulo `re`: findall, match, search, split, sub

En este notebook aprender√°s las **funciones principales del m√≥dulo `re`** y sus diferencias pr√°cticas.

Conocerlas a fondo te permitir√° elegir la adecuada seg√∫n la tarea: **buscar, extraer, separar o reemplazar texto.**

````python
import re
print('‚úÖ M√≥dulo re cargado correctamente.')
````

## 1Ô∏è‚É£ `findall()` ‚Äì Buscar todas las coincidencias

Devuelve una **lista** con todas las apariciones del patr√≥n.

**Ejemplo:** extraer todos los n√∫meros de una cadena.

````python
texto = 'El producto 123 cuesta 45 euros y el env√≠o 5 euros.'
numeros = re.findall(r'\d+', texto)
print(numeros)
````

## 2Ô∏è‚É£ `match()` ‚Äì Coincidencia al inicio

Comprueba si el texto **empieza** con el patr√≥n.

Devuelve un objeto `Match` si coincide, o `None` si no.

````python
texto = 'Python es genial'
inicio = re.match(r'Python', texto)
print('Coincidencia:', inicio)
if inicio:
    print('‚úîÔ∏è Empieza con:', inicio.group())
````

## 3Ô∏è‚É£ `search()` ‚Äì Primera coincidencia en cualquier lugar

Busca la **primera aparici√≥n** del patr√≥n en toda la cadena.

````python
texto = 'Aprender Python es divertido'
coincidencia = re.search(r'Python', texto)
if coincidencia:
    print('Encontrado en posici√≥n:', coincidencia.start())
````

## 4Ô∏è‚É£ `split()` ‚Äì Dividir texto por patr√≥n

Permite dividir una cadena seg√∫n un separador (regex).

````python
texto = 'uno, dos; tres , cuatro:palabra'
partes = re.split(r'[:;, ]+', texto)
print(partes)
# Devuelve un objeto de tipo lista con las partes obtenidas tras la divisi√≥n
````

## 5Ô∏è‚É£ `sub()` ‚Äì Buscar y reemplazar

Permite reemplazar texto que coincida con un patr√≥n.

**Ejemplo:** reemplazar n√∫meros por `#`.

````python
texto = 'Tel√©fono: 123-456-789'
resultado = re.sub(r'\d', '#', texto)
print(resultado)
# re.sub() devuelve una nueva cadena de texto con las sustituciones ya realizadas
````

## üß† Ejercicio guiado

1. Dado el texto:
   ```python
   'Mi usuario es david2025 y mi backup es david2024'
   ```
   - Usa `findall()` para extraer los a√±os.
   - Usa `sub()` para reemplazar los n√∫meros por `[A√ëO]`.

2. Comprueba si el texto **empieza por 'Mi'** usando `match()`.

````python
# ‚úèÔ∏è Escribe aqu√≠ tu c√≥digo
texto = 'Mi usuario es david2025 y mi backup es david2024'
# findall
anios = re.findall(r'\d+', texto)
print('A√±os encontrados:', anios)
print('-'*50)

# sub
sustitucion = re.sub(r'\d{4}', '[A√ëO]', texto)
print('Texto reemplazado:', sustitucion)
print('-'*50)

# match
coincidencia = re.match(r'Mi', texto)
if coincidencia:
    print('‚úÖ, el texto comienza por "Mi"')
    print('-'*50)
else:
    print('El texto no empieza por "Mi"')
    print('-'*50)
````

## ‚úÖ Soluci√≥n propuesta

````python
texto = 'Mi usuario es david2025 y mi backup es david2024'

anios = re.findall(r'\d{4}', texto)
print('A√±os encontrados:', anios)

nuevo = re.sub(r'\d{4}', '[A√ëO]', texto)
print('Texto reemplazado:', nuevo)

if re.match(r'Mi', texto):
    print('‚úîÔ∏è Empieza con "Mi"')
````

---
**Resumen:**

| Funci√≥n | Descripci√≥n | Devuelve |
|:---------|:-------------|:----------|
| `findall()` | Todas las coincidencias | Lista |
| `match()` | Solo al inicio del texto | Objeto `Match` o `None` |
| `search()` | Primera coincidencia | Objeto `Match` o `None` |
| `split()` | Divide texto por patr√≥n | Lista de partes |
| `sub()` | Reemplaza coincidencias | Nuevo texto |

Estas funciones son el **n√∫cleo pr√°ctico del m√≥dulo `re`**. Las usar√°s constantemente en validaciones, limpieza y procesamiento de texto.
