```python
# ASOCIACIÓN -> relación débil -> dos objetos se conocen, pero ninguno pertenece al otro, ni depende
# de su ciclo de vida.

class Profesor:
    def __init__(self, nombre):
        self.nombre = nombre

    def enseñar(self, alumno):
        print(f'{self.nombre} enseña a {alumno.nombre}') # utilizará la futura clase Alumno

class Alumno:
    def __init__(self, nombre):
        self.nombre = nombre

prof = Profesor('Yoni')
alum = Profesor('Felipe')

prof.enseñar(alum)
```

```python
# AGREGACIÓN -> relación débil pero con contenedor. Un objeto contiene a otro, pero si el contenedor
# muere, los contenidos pueden seguir existiendo

# Ej real: un equipo tiene jugadores, pero si se borra el equipo, los jugadores siguen existiendo
# individualmente

class Jugador:
    def __init__(self, nombre):
        self.nombre = nombre

class Equipo:
    def __init__(self, nombre):
        self.nombre = nombre
        self.jugadores = []

    def añadir_jugador(self, jugador):
        self.jugadores.append(jugador.nombre)

    def mostrar_equipo(self):
        print(self.jugadores)


atm = Equipo('Atlético de Madrid')
giuliano = Jugador('Giuliano Simeone')
julian = Jugador('Julián Álvarez')

atm.añadir_jugador(giuliano)
atm.añadir_jugador(julian)

atm.mostrar_equipo()
```

```python
# COMPOSICIÓN -> relación más fuerte -> un objeto contiene a otros que no pueden existir sin él.

# EJ: Un coche está compuesto por motor, ruedas, chasis... si destruyes el coche, se destruye todo

class Motor:
    def __init__(self, tipo):
        self.tipo = tipo

class Chasis:
    def __init__(self, tipo):
        self.tipo = tipo

class Rueda:
    def __init__(self, tipo):
        self.tipo = tipo

class Coche:
    def __init__(self, marca, motor, chasis, rueda):
        self.marca = marca
        self.motor = motor
        self.chasis = chasis
        self.rueda = rueda

    def mostrar_info(self):
        print(f'El coche es de marca {self.marca}, tiene un motor {self.motor.tipo} y usa unas ruedas {self.rueda.tipo}.')

coche1 = Coche('Renault', motor= Motor('V8'), chasis= Chasis('Monocasco'), rueda= Rueda('redondas'))
coche1.mostrar_info()
```

El código anterior está bien pero no es composición. Es agregación. Para ser composición la parte tiene que crearse dentro del otro objeto, para que dependa completamente de él.

```python
class Motor:
    def __init__(self, tipo):
        self.tipo = tipo

class Coche:
    def __init__(self, marca):
        self.marca = marca
        self.motor = Motor('V8')

    def mostrar_info(self):
        print(f'Coche {self.marca} creado con motor {self.motor.tipo}')

coche1 = Coche('Renault')
coche1.mostrar_info()
```

```python
# EXPERIMENTAR!!

from datetime import datetime
from collections import defaultdict
import copy

class Producto:
    def __init__(self, nombre, precio):
        self.nombre = nombre
        self.precio = precio

class Cliente:
    def __init__(self, id):
        self.id = id
        self.carrito = Carrito() # Composición

class Carrito:
    def __init__(self):
        self.productos = defaultdict(int) # cada clave empieza con un valor 0.
        # NOTA: Carrito tiene objetos Producto, no solo sus nombres.
    
    def añadir_producto(self, producto):
        self.productos[producto] += 1 # Agregación cuando el producto de .append(producto) sea un Producto

    def quitar_producto(self, producto):
        if producto not in self.productos:
            print("Producto no encontrado en el carrito")
            return

        if self.productos[producto] > 1:
            self.productos[producto] -= 1
        else:
            del self.productos[producto]


    def mostrar_carrito(self):
        return {k.nombre: v for k, v in self.productos.items()}
    
class Orden:
    def __init__(self, carrito):
        self.fecha = datetime.now().strftime("%d/%m/%Y %H:%M:%S")
        self.cliente = None
        self.productos = copy.deepcopy(carrito.productos) # Se hace una copia para que la orden
        # tenga los mismos productos que el carrito en el momento de crear la orden y no varíe 
        # cuando variemos el carrito después.
        self.total = self.calcular_total()

    def mostrar_productos(self):
        return {k.nombre : v for k,v in self.productos.items()}

    def mostrar_fecha_orden(self):
        return f'Orden realizada en: {self.fecha}'
    
    def calcular_total(self):
        return sum([k.precio*v for k,v in self.productos.items()])
    
    def mostrar_detalle(self):
        cliente_id = self.cliente.id if self.cliente else "Desconocido"
        print(f"Orden del cliente ID: {cliente_id}")
        print(f"Fecha: {self.fecha}")
        print("Productos:")
        for producto, cantidad in self.productos.items():
            print(f"  - {producto.nombre} x{cantidad} ({producto.precio}€ cada uno)")
        print(f"Total: {self.total}€")
```

```python
carr1 = Carrito()
carr1.añadir_producto(producto= prod1)
carr1.mostrar_carrito()
```
